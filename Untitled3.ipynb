```python
# Data Cleaning
import pandas as pd
import re

# Charger les données à partir du fichier CSV 
data = pd.read_csv('imdb_comments.csv')

# Supprimer les balises HTML des commentaires
data['Commentaire'] = data['Commentaire'].apply(lambda x: re.sub('<.*?>', '', x))

# Supprimer les caractères spéciaux et la ponctuation
data['Commentaire'] = data['Commentaire'].apply(lambda x: re.sub('[^a-zA-Z0-9\s]', '', x))

# Mettre en minuscules
data['Commentaire'] = data['Commentaire'].str.lower()

# Supprimer les espaces vides supplémentaires
data['Commentaire'] = data['Commentaire'].apply(lambda x: ' '.join(x.split()))

# Afficher les premières lignes des données nettoyées
print(data.head())
```

```python
from textblob import TextBlob
import pandas as pd

# Charger les données à partir du fichier CSV (ajustez le chemin du fichier)
imdb_comments = pd.read_csv('imdb_comments.csv')

# Définir une fonction pour étiqueter le sentiment en utilisant TextBlob
def etiqueter_sentiment(Commentaire):
    analysis = TextBlob(Commentaire)
    sentiment = analysis.sentiment.polarity

    if sentiment > 0:
        return "positif"
    elif sentiment < 0:
        return "négatif"
    else:
        return "neutre"

# Appliquer la fonction d'étiquetage pour créer une nouvelle colonne "sentiment" dans imdb_comments
imdb_comments['sentiment'] = imdb_comments['Commentaire'].apply(etiqueter_sentiment)

# Afficher les premières lignes des données avec les étiquettes de sentiment
print(imdb_comments[['Commentaire', 'sentiment']].head())
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import f1_score
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler
from imblearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import accuracy_score, confusion_matrix


# Charger vos données
imdb_comments = pd.read_csv('imdb_comments.csv')

# Divisez l'ensemble de données en caractéristiques (X) et étiquettes (y)
X = imdb_comments['Commentaire']
y = imdb_comments['sentiment']

# Divisez l'ensemble de données équilibré en un ensemble d'entraînement (70%) et un ensemble de test (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Créez un objet TfidfVectorizer pour convertir le texte en vecteurs de caractéristiques
tfidf_vectorizer = TfidfVectorizer()

# Ajustez le vectoriseur sur les données d'entraînement et transformez-les
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Définissez les taux de suréchantillonnage et de sous-échantillonnage
over_sampling_ratio = 0.8
under_sampling_ratio = 0.7

# Définissez les taux de suréchantillonnage et de sous-échantillonnage
sampling_strategy_over = {
   'positif': 11850,
    'négatif': 9000,
    'neutre': 9000
}
sampling_strategy_under = {
   'positif': 7500,
    'négatif': 7000,
    'neutre':6500
}

# Créez un objet RandomOverSampler
over_sampler = RandomOverSampler(sampling_strategy=sampling_strategy_over, random_state=42)

# Appliquez le suréchantillonnage
X_over, y_over = over_sampler.fit_resample(X_train_tfidf, y_train)

# Créez un objet RandomUnderSampler
under_sampler = RandomUnderSampler(sampling_strategy=sampling_strategy_under, random_state=42)

# Appliquez le sous-échantillonnage
X_balanced, y_balanced = under_sampler.fit_resample(X_over, y_over)

# Créez un pipeline avec Random Oversampling, Random Undersampling et un classificateur DecisionTree
pipeline = Pipeline([
    ('model', DecisionTreeClassifier(random_state=42))
])

# Entraînez le modèle sur l'ensemble d'entraînement prétraité
pipeline.fit(X_balanced, y_balanced)

# Prétraitement des données de test
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Prédisez les étiquettes sur l'ensemble de test
y_pred = pipeline.predict(X_test_tfidf)

# Calculez l'accuracy
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy sur l'ensemble de test:", accuracy)

# Générez la matrice de confusion
confusion = confusion_matrix(y_test, y_pred)

# Étiquettes possibles
labels = ['positif', 'négatif', 'neutre']

# Affichez la matrice de confusion avec les étiquettes
print("Matrice de confusion:")
print(confusion)
```
